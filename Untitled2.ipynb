```
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
import torch
```

```
2e-3
```

```
N=100
D=2
K=3
X=np.zeros((N*K,D))
y=np.zeros(N*K,dtype='uint8')
```

```
for j in range(K):
  ix=range(N*j,N*(j+1))
  r=np.linspace(0.0,1,N)
  t=np.linspace(j*4,(j+1)*4,N)+np.random.randn(N)*0.2
  X[ix]=np.c_[r*np.sin(t),r*np.cos(t)]
  y[ix]=j
plt.figure(figsize=(10,8))
plt.scatter(X[:,0],X[:,1],c=y,s=40,cmap=plt.cm.brg)
plt.show
```

```

X=torch.autograd.Variable(torch.FloatTensor(X))
y=torch.autograd.Variable(torch.LongTensor(y.astype(np.int64)))
```

```
y=y.unsqueeze(dim=1)
y.data.shape
```

```
ndim_in=2
ndim_out=3
num_epochs=500
learning_rate=2e-2

neuron= torch.nn.Sequential(
    torch.nn.Linear(ndim_in, ndim_out)
    
   
    
)


loss_fn=torch.nn.CrossEntropyLoss(size_average=False)
#loss_fn=torch.nn.MSELoss(size_average=False)

optimizer=torch.optim.SGD(neuron.parameters(),lr=learning_rate)


for i in range(num_epochs):
  y_pred=neuron(X)
  loss = loss_fn(y_pred,torch.max(y,1)[0])
  #print(f'Iteration:{i}->loss:{loss.data}')
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()
```

```
X = X.data.numpy()
y = y.data.numpy()
```

```
h = 0.02
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
 
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))
grid_tensor = torch.FloatTensor(np.c_[xx.ravel(), yy.ravel()])
 
Z = neuron(torch.autograd.Variable(grid_tensor))
Z = Z.data.numpy()
Z = np.argmax(Z, axis=1)
Z = Z.reshape(xx.shape)
 
plt.figure(figsize=(10, 8))
 
plt.contourf(xx, yy, Z, cmap=plt.cm.brg, alpha=0.3)
plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap=plt.cm.brg)
 
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
 
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$y$', fontsize=14)
plt.show();
```
